# Exploratory Data Analysis (EDA)

```python
ax = sns.countplot(x="bedrooms", data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

from above graph it can be concluded that most of the houses are of 3 bedrooms. 

```python
ax = sns.countplot(x="floors", data=df)

sns.set(rc={'figure.figsize':(10,5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

from above graph it can be concluded that most of the houses are of sigle storey houses. 

```python
price_check = df.groupby(["floors"], as_index=False)["price"].sum().sort_values(by=["price"], ascending=False)

sns.set(rc={'figure.figsize':(8,5)})

sns.barplot(data = price_check, x="floors", y="price")
```

from above graph it can be concluded that the price of 2 storey houses are high. 

```python
price_check = df.groupby(["condition"], as_index=False)["price"].sum().sort_values(by=["price"], ascending=False)

sns.set(rc={'figure.figsize':(8,5)})

sns.barplot(data = price_check, x="condition", y="price")
```

from above graph it can be concluded that the price of houses varies with the condition of house. 

```python
price_check = df.groupby(["yr_built"], as_index=False)["price"].sum().sort_values(by=["price"], ascending=False).head(10)

sns.set(rc={'figure.figsize':(10,5)})

sns.barplot(data = price_check, x="yr_built", y="price")
```

from above graph it can be concluded that the houses that build in 2006 have higher prices. 

```python
price_check = df.groupby(["city"], as_index=False)["price"].sum().sort_values(by=["price"], ascending=False).head(10)

sns.set(rc={'figure.figsize':(10,5)})

sns.barplot(data = price_check, x="city", y="price")
```

from above graph it can be concluded that the seattle houses price are higher amoung all. 

```python
price_check = df.groupby(["date"], as_index=False)["price"].sum().sort_values(by=["price"], ascending=False).head()

sns.set(rc={'figure.figsize':(10,5)})

sns.barplot(data = price_check, x="date", y="price")
```

from above graph it can be concluded that the houses that are build in 6 month of 2014 have high prices.
